# 获取页面

```python
import requests
```

```python
!curl -v "https://movie.douban.com/subject/33404425/comments?start=0&limit=20&sort=new_score&status=P&percent_type=h"
```

```python
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/83.0.4103.116 Safari/537.36"
}
def  get_page(url):
    r = requests.get(url, headers=headers)
    return r.text
    
page = get_page('https://movie.douban.com/subject/33404425/comments?start=0&limit=20&sort=new_score&status=P&percent_type=h')
```

# 解析页面
https://www.crummy.com/software/BeautifulSoup/bs4/doc.zh/

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(page, 'html.parser')

print(soup.prettify())
```

```python
soup.select('span[class="short"]')
```

```python
[node.text for node in soup.select('span[class="short"]')]
```

```python
def parse(page):
    soup = BeautifulSoup(page, 'html.parser')
    return [node.text for node in soup.select('span[class="short"]')]
```

# 爬取1000个好评

```python
import time
```

```python
url = 'https://movie.douban.com/subject/33404425/comments?start={}&limit=20&sort=new_score&status=P&percent_type=h'
result = []
for i in range(0, 10):
    page = get_page(url.format(i*20))
    comments = parse(page)
    result.extend(comments)
    time.sleep(3)
    print(i,len(comments), len(result))
good_result = result
```

```python
import pandas as pd 
pd.DataFrame(good_result)
```

# 爬取200条差评

```python
url = 'https://movie.douban.com/subject/33404425/comments?start={}&limit=20&sort=new_score&status=P&percent_type=l'
result = []
for i in range(0, 10):
    page = get_page(url.format(i*20))
    comments = parse(page)
    result.extend(comments)
    time.sleep(3)
    print(i,len(comments), len(result))
bad_result=result
```

# 用pandas 保存文件

```python
import pandas as pd
```

```python
good = pd.DataFrame(good_result)
good[1]=1
bad = pd.DataFrame(bad_result)
bad[1]=0
result = pd.concat([good,bad], axis=0)
result.columns=['text', 'label']
result
```

```python
result.to_csv("隐秘的角落评论.csv", index=False)
```
